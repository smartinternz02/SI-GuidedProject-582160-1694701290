```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('House Price India.csv')
df=df.head(20)
df
```

### Univariate Analysis

```python
plt.plot(df['number of bedrooms'])
```

```python
plt.plot(df['number of floors'])
```

### Bivariate Analysis

```python
plt.plot(df['number of views'],df['number of bedrooms'],'o')
plt.title('No of bedroom vs No of views')
plt.xlabel('No of views')
plt.ylabel('No of bedroom')
```

```python
sns.boxplot(data=df, x='number of bedrooms', y='Price')
plt.title('Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

### Multivariate Analysis

```python
numerical_vars = ['number of bedrooms', 'number of bathrooms', 'living area', 'lot area', 'Number of schools nearby', 'Distance from the airport', 'Price']
correlation_matrix = df[numerical_vars].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
